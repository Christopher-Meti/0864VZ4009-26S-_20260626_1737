### (1) Module RE und seine Funktionen

Überblick über die wichtigsten Funktionen für das Aufsuchen und Matchen von Mustern, die als Regular Expressions formuliert werden können, in einem String

|Funktion        |Erklärung|
|:---------------|:--------|
|re.search(A, B)|Auffinden des Muster in A in String B, Ausgabe des ersten Matches|
|re.match(A, B)|Auffinden des Musters A am Anfang von String B|
|re.fullmatch(A, B)|Muster A entspricht dem String B|
|re.findall(A, B)|Auffinden von allen (nicht-überlappenden= Muster A in String B, Ausgabe als Liste|
|re.split(A, B)|Aufteilen des Strings B an der Stelle, wo Muster A vorkommt|
|re.sub(A, B, C)|Ersetzen aller Muster A in String C durch Muster B, Rückgabe: veränderter String|
|re.subn(A, B, C)|Ersetzen aller Muster A in String C durch Muster B, Rückgabe: tuple mit neuem String & Anzahl an Ersetzungen|
|re.escape(A)|Sonderzeichen werden im Muster A mit Escape-Zeichen versehen|

````python
# FULLMATCH
fullmatch1 = re.fullmatch(patternA, stringB)
print(fullmatch1)
re.fullmatch(stringB, stringB)
````

### (2) Regular-Expressions-Objekte: re.compile()
Wenn man eine Regular Expression öfters braucht, dann macht es Sinn, dieses als eigenes Objekt zu speichern. Im Modul _re_ kann ein solches Objekt mithilfe der `re.compile(pattern, flags=0)` erzeugt werden. 

PARAMETER
- `pattern` entspricht einer Regular Expressionerzeugt ein wiederverwendbares Pattern‑Objekt. Üblicherweise wird ein ```r"…"``` vorangesetzt, damit Backslashes nicht als Python‑Escapes interpretiert werden.
- `flag` entspricht den Optionen, die zusätzlich gesetzt werden können (wie ```re.IGNORECASE```, ```re.MULTILINE```, ```re.UNICODE``` etc.). Mehrere Flags werden mit `OR`oder `|` kombiniert.

````python
# Erzeugen eines Regular-Expressions-Objekts
regEx = re.compile(pattern = "(ei)n(es)?")
print(regEx)
print(type(regEx))
````

Ein Regular-Expressions-Objekt kann alle die obigen _re-Funktionen_ nutzen, zusätzlich gibt es ein paar weitere Methoden: 

- `.flags`: Ausgabe der gesetzten Flags plus den Inline-Flags (wie (?...)) und impliziten Flags wie UNICODE. 
- `.groups`: Anzahl der erfassenden Gruppen im Pattern (bei Verwendung von `()`. 
- `.pattern`: Ausgabe des Regular Expressions 

````python
# Anwenden einer Funktion
regEx.findall(stringB)
````

````python
regEx.flags
````

````python
regEx.groups
````

````python
regEx.pattern
````

### (2) Flags gezielt einsetzen

Wichtige Flags:
- ``re.I`` (ignorecase): Groß-/Kleinschreibung ignorieren
- ``re.M`` (multiline): ^ und $ verankern an Zeilenanfang/-ende, nicht nur Datei
- ``re.S`` (dotall): Punkt . matcht auch Newline
- ``re.X`` (verbose): zur Vereinfachung von Regular Expressions: Mit diesem Flag können diese übersichtlicher und lesbarer gestaltet werden (z. B. durch eine visuelle Trennung der einzelnen Abschnitte) 

````python
print(stringB)
````

````python
# IGNORECASE: Was passiert, wenn man nach "beispiel" sucht?
re.findall(pattern=r"beispiel", string=stringB, flags=re.I)
re.findall(pattern=r"[bB]eispiel", string=stringB)
````

````python
# ASCII-Standard: Was passiert, wenn man nach Wörter im ASCII-Code sucht?
re.findall(r"\w+", stringB)
re.findall(r"\w+", stringB, flags=re.A)
````

````python
# MULTILINE: Was passiert bei Zeilenumbrüchen, wenn man nach dem ersten Wort sucht?
stringC = "Das ist ein Satz,\ndas ist ein neuer Satz"
re.findall(r"^\w+", stringC)
re.findall(r"^\w+", stringC, flags=re.M)
````

````python
# VERBOSE: Wie kann eine komplexere Regular Expression lesbarer gestaltet werden?
phrases1 = re.findall(r"für\s+(.+?)(?:,|$)", stringB)
phrases2 = re.findall(pattern=r"""
für
\s+   # mind. ein Leerzeichen
(.+?)
(?:,|$)
""", string=stringB, flags=re.X)

print(phrases1)
print(phrases2)  
````

````python
re.findall(r"für\s+(.+?)(?:,|$)", stringB)
````

````python
print(stringB)
````

````python
# Kombination von Flags
re.findall(pattern=r"""
^.*
""", string=stringC, flags=re.M|re.X)
````

(B) Wenden Sie folgenden Aufgaben auf den Textteil an:
1. Entfernen aller Sonderzeichen wie " * = -> etc.
2. Entfernen aller Nennungen, die mit @ beginnen 
3. Entfernen aller #-Tags
4. Entfernen aller URLS
6. Ersetzen aller Mojibacks (Ã¼ = ü, Ã¶ = ä, Ã¤ = ö, ÃŸ = ß, Ãœ = Ü, Ã– = Ö, Ã„ = Ä) durch korrekte Zeichen
7. Ersetzen aller HTML-Entitäten (&uuml;=ü, &Uuml;=Ü, &auml;=ä, &Auml;=Ä, &ouml;=ö, &Ouml;=Ö, &szlig;=ß, &amp;=&, &quot;=", &apos;=', &lt;=<, &gt;=>) durch korrekte Zeichen
8. Entfernen einiger typischen Tippfehler

````python
# Posting zur Bereinigung
post = """ID: R-0001 | Timestamp: 2025-01-12 09:41:22 | Channel: Web | Rating: 2 | Status: open
Tags: lieferung, paket, verpakung, verspätung
Text:
Guten Tag***, mein Paket ist leider noch nich angekommen 😕  .
Tracking hängt seit 5 Tagen ==> auf "im Verteilzentrum". Bitte pr&uuml;fen, danke!
Hab extra EXPRESS gebucht!!!  Sie haben wierklich  keine Infos geliefert. Ã¼blich
#Service?? #fail Mehr Details: https://bit.ly/3xyz  bzw. http://example.com/path?a=1&b=2
Kontakt: @ShopSupport*  Ersatz bitte seperat senden -> danke! ==< >  E-Mail: info@example.com
"""

# Hilfsfunktion für die Ausgabe von Patterns
def show(title, value):
    print(f"\n=== {title} ===")
    print(value)
````

````python
# Testen der Hilfsfunktion
show("Rohtext", post)
````

````python
# (A) Extrahieren von Text & Metadaten
head, text = re.split("Text:", post, 1)
show("Textteil", text)
````

````python
# 1 Entfernen aller Sonderzeichen wie " * = -> etc.
sonderzeichen = re.compile(r"[=<>\*&]+")
text1 = sonderzeichen.sub(" ", text)
show("Titel", text1)
````

````python
# 2 Finde alle Nennungen, die mit @ beginnen und entferne diese
nennungen = re.compile(r"@[A-Za-z0-9_.-]+\*?")
text1 = nennungen.sub("", text1)
show("(2) @-Nennungen entfernt", text1)
````

````python
# 3 Entfernen aller #-Tags
hashtags = re.compile(r"(?:^|\s)#[A-Za-z0-9_ÄÖÜäöüß]+(?=$|\s|[^\wÄÖÜäöüß])")
text1 = hashtags.sub(" ", text1)
show("(3) #-Tags entfernt", text1)
````

````python
# 4 Entfernen aller URLS
web_urls = re.compile(r"https?://\S+")
text1 = web_urls.sub("", text1)
show("(4) URLs entfernt", text1)
````

````python
# 5 Entfernen von Mojibacks (mithilfe von String-Operationen)
text1 = (text1
          .replace("Ã¼", "ü")
          .replace("Ã¶", "ö")
          .replace("Ã¤", "ä")
          .replace("ÃŸ", "ß")
          .replace("Ãœ", "Ü")
          .replace("Ã–", "Ö")
          .replace("Ã„", "Ä"))

show("(5) Mojiback entfernt", text1)
````

````python
# 6 Entfernen von Entitäten
text1 = (text1
          .replace("&uuml;", "ü").replace("&Uuml;", "Ü")
          .replace("&auml;", "ä").replace("&Auml;", "Ä")
          .replace("&ouml;", "ö").replace("&Ouml;", "Ö")
          .replace("&szlig;", "ß")
          .replace("&amp;", "&").replace("&quot;", '"')
          .replace("&apos;", "'").replace("&lt;", "<").replace("&gt;", ">"))

show("(6) Entitäten entfernt", text1)
````

````python
# 7 Entfernen von typischen Tippfehlern
text1 = re.sub(r"\bnich\b",        "nicht",      text1, flags=re.I)
text1 = re.sub(r"\bwierklich\b",   "wirklich",   text1, flags=re.I)
text1 = re.sub(r"\bverpakung\b",   "verpackung", text1, flags=re.I)
text1 = re.sub(r"\bseperat\b",     "separat",    text1, flags=re.I)
text1 = re.sub(r"\bprufen\b",      "prüfen",     text1, flags=re.I)

show("(7) Entfernen von typischen Tippfehlern", text1)
````

````python
# 8 Entfernen von Tabulatoren, mehreren Zeilenumbrüchen (= Leerzeilen)
tabulatoren = re.compile(r"[ \t]+")
text1 = tabulatoren.sub(" ", text1)
show("(9a) Tabulatoren entfernt", text1)

leerzeilen = re.compile(r"\n{3,}")
text1 = leerzeilen.sub(" ", text1)
show("(9b) Leerzeilen entfernt", text1)
````

````python
# 9 Entfernen von Mehrfachleerzeichen
text1 = text1.strip()
show("(10) Tabulatoren entfernt", text1)
````
