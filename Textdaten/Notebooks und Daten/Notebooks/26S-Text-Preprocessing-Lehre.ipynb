#### **Entfernen von zusätzlichen Zeichen**
In einem Text können Zeichen vorhanden sein, die für die NLP-Modellierung nicht notwendig sind:
* _Satzzeichen_ stehen direkt hinter einem Wort (ohne Abstand), daher kann es passieren, dass sie bei der Tokenisierung als ein Teil des Wortes gesehen werden. Für ein Topic-Modeling sind Satzzeichen (und andere Zeichen) nicht notwendig, daher werden sie hier entfernt.
* _Emojis_ sind in einigen Anwendungen evtl. notwendig, z. B. bei Sentimentanalysen. Für Topic-Modeling braucht man diese üblicherweies nicht. 
* _Symbole_ wie aus Mathematik, Physik. Hier muss man eine Entscheidung treffen, ob diese notwendig sind.
  
Genutzt werden folgende Funktionen aus der Library *string*
* `str.maketrans(x,y,z)`: Erzeugen einer Zuordnungstabelle/eines Wörterbuchs: `x`, `y`= gleich lange Zeichenketten; Stellvertreter für String mit Zeichen zu ersetzen bzw. Ersatzzeichen, `z` = Zeichenkette, gibt an, welche Zeichen zu ersetzen sind;
* `punctuation()`: Satzzeichen (aus ASCII-Code) !"#$%&'()*+,-./:;<=>?@[\]^_{ 
* `translate()`: ersetzt bestimmte Zeichen durch andere Zeichen; Ersetzungen müssen vorher definiert werden, z. B. über Zuordnungstabelle oder Wörterbuch

```python
# Entfernen von häufigen nicht-ASCII-Satzzeichen
# (z. B. deutsche Anführungszeichen, Gedankenstriche, Ellipse)
extra_punct = '„“‚’«»–—…·•‚‒‹›'
text_nopunct = text_nopunct.translate(str.maketrans('', '', extra_punct))
print(text_nopunct)
```

##### Entfernen von anderen Zeichen

* `isalpha()`: True, wenn ein Zeichen ein Buchstabe ist (a-z)
* `isdigit()`: True, wenn ein Zeichen eine Ziffer ist (0-9)
* `isspace()`: : True, wenn ein Zeichen ein Leerzeichen ist 

```python
# Entfernen von Emojis und sonstigen Symbolen; es sollen nur Buchstaben, Ziffern und Leerzeichen erhalten bleiben; 
# (-> dadurch verschwinden Emojis, Währungs-/Mathe-Symbole etc.)
clean_chars = []
for ch in text_nopunct:
    if ch.isalpha() or ch.isdigit() or ch.isspace():
        clean_chars.append(ch)
    # alles andere (Emojis/Symbole) wird verworfen

text_clean = ''.join(clean_chars)

print(text_clean)
```

### **Schritt 2: Tokenisierung**

Bei einer manuellen Tokenisierung muss man auf die Zeichenebene gehen. Eine Möglichkeit besteht darin, Wörter als Tokens zu nehmen, da Leerzeichen die Grenzen für diese Tokens darstellen. D.h. man durchsucht einen String auf Zeichenebene und findet die einzelnen Tokens, die zwischen Leerzeichen stehen. 
- Wir zerlegen den Text in „Wort“-Tokens
 - Regel: Nur Buchstaben (a–z, A–Z und Umlaute/ß werden von isalpha() als Buchstaben erkannt)
 - Alles Nicht-Alphabetische (Zahlen, Emojis, Satzzeichen) trennt Tokens

```python
# Tokenisierung über Zeichen & Schleife

filtered_chars = []
for ch in text_nopunct:
    if ch.isalpha() or ch.isdigit() or ch.isspace():
        filtered_chars.append(ch)
    else:
        filtered_chars.append(' ')

filtered_text = ''.join(filtered_chars)

# 2) Tokens erzeugen: Wörter oder Zahlen
tokens = [tok for tok in filtered_text.split() if tok.isalpha() or tok.isdigit()]

print(tokens)
```

```python
# Tokenisierung über split & Leerzeichen
# Voraussetzung: Sonderzeichen vorhanden

raw_tokens = text_nopunct.split()

# Token ist ein Wort (nur Buchstaben) oder eine Zahl (nur Ziffern)
tokens = [tok for tok in raw_tokens if tok.isalpha() or tok.isdigit()]

print(tokens)
```

### **Schritt 3: Filtern der Tokens**

#### **Entfernen von Stoppwörter**

Was sind Stoppwörter? Stoppwörter sind Wörter, die häufig auftreten, aber wenig Information enhalten wie 
- Pronomen wie *ich*, *du*, *wir*, ...
- Artikel wie *der*, *eine*, ...
- Konjunktionen wie *und*, *oder*, *aber*, ...
- Hilfsverben wie *ist*, *war*, ...
- Füllwörter (kontextabhängig) wie *bitte*, *hallo* ...
- Präpositionen wie *zu*, *bei*,  ...
- ...

Gerade für Tasks wie Topic-Modeling braucht es solche Funktionswörter nicht, da es um lexikalische Morpheme geht. 

Hinweise:
   * Für Topic Modeling entfernt man oft Funktionswörter (Artikel, Pronomen, Hilfsverben, Präpositionen)
   * Ob Negationen wie „nicht“ entfernt werden sollen, hängt vom Ziel ab (für Sentiment eher behalten)

```python
# Erstellen einer Liste an Stoppwörter
stopwords_demo = {
    # Artikel/Pronomen
    "der","die","das","ein","eine","einer","eines","einem","einen",
    "ich","du","er","sie","es","wir","ihr","ihnen","ihre","ihrem","meine","mein","dein","sein","ihr",
    # Konjunktionen/Partikeln
    "und","oder","aber","denn","doch","jedoch",
    # Präpositionen/Adverbien
    "zu","zum","zur","im","in","ins","am","an","auf","aus","bei","mit","ohne","für","von","vor","nach","über",
    # Hilfs-/Modalverben
    "ist","sind","war","waren","bin","bist","seid","sein","wird","werden",
    # Höflichkeit/Füllwörter (kontextabhängig)
    "bitte","danke","leider","noch","guten","tag","hallo"
}
print(stopwords_demo)
```

```python
# Optional: Negationswörter (für Topic Modeling oft entfernt; zum Zeigen beides vorbereiten)
negations = {"nicht","kein","keine","keinen","keinem","keiner","keines"}
print(negations)
```

```python
# Stopwörter entfernen (Variante A: Negation wird entfernt)
tokens_no_stop_remove_neg = [
    t for t in tokens
    if (t not in stopwords_demo) and (t not in negations)
]

print(tokens_no_stop_remove_neg)
```

```python
# Stopwörter entfernen (Variante B: Negation bleibt erhalten)
tokens_no_stop_keep_neg = [
    t for t in tokens
    if (t not in stopwords_demo) or (t in negations)  # Negationen ausdrücklich zulassen
]

print(tokens_no_stop_keep_neg)
```

##### Tokens filtern
- Entfernt sehr kurze Tokens, z. B. Länge < 3
- Ziel: Rauschen reduzieren (v. a. bei Abkürzungen o. Ä.)

#### **Entfernen von kurzen Tokens**

```python
# Entfernung von Tokens mit geringer Länge
min_len = 3
tokens_len3 = [t for t in tokens_no_stop_keep_neg if len(t) >= min_len]

print(tokens_len3)
```

```python
# aufbereiteter Text
reconstructed_remove_neg = " ".join(tokens_no_stop_remove_neg)
reconstructed_keep_neg   = " ".join(tokens_len3)

print("Variante A (Negation entfernt):")
print(reconstructed_remove_neg)
print("\nVariante B (Negation behalten, Länge >= 3):")
print(reconstructed_keep_neg)
```

## **Teil 2: Preprocessing mit Libraries**

In Python gibt es Libraries fürs Preprocessing von Textdaten, insbesondere für die Vorbereitung auf NLP-Aufgaben:
- NLTK
- spaCy
- scikit-learn
- Gensim
- Stanza
- Flair/Transformers (Hugging Face)

**NLTK** enthält viele Bausteine, die einzeln aufgerufen werden kann. Damit können die einzelnen Schritte besser nachvollzogen werden. 
Nachteil: Die Lemmatisierung auf Deutsch ist eher schwächer, für englischsprachige Texte besser geeignet. Die Qualität ist auch nicht gleichbleibend. 

**spaCY** ist sehr schnell und leicht erweiterbar; kann bei Tokenisierung und Lemmatisierung mit deutschen Texten gut umgehen und erkennt Emojis, URLS, Satzzichen etc. 
Nachteil: Es ist kein Stemming möglich und die Modelle müssen separat geladen werden. 

**scikit-learn** ist sehr praktikabel, um Features in Texten zu finden. 
Nachteil: Keine Lemmatisierung möglich, die Tokenisierung ist auch sehr einfach. 



#### **Minibeispiele für den Vergleich**

```python
text = "Guten Tag, mein Paket ist leider noch nicht angekommen."
```

##### **1 Tokenisierung/Stemming mit NLTK**

```python
# Tokenisierung
import nltk
from nltk.tokenize import wordpunct_tokenize

tokens_nltk = nltk.word_tokenize(text)
print(tokens_nltk)
```

```python
# Lemmatisierung
from nltk.stem import WordNetLemmatizer
#nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

tokens_lemma = [lemmatizer.lemmatize(word) for word in tokens_nltk]
print(tokens_lemma)
```

```python
# Stemming
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import PorterStemmer

stemmer_sn = SnowballStemmer("german")   
tokens_snowball = [stemmer_sn.stem(t) for t in tokens_nltk]
print(tokens_snowball)

stemmer_ps = PorterStemmer()
tokens_porter = [stemmer_ps.stem(t) for t in tokens_nltk]
print(tokens_porter)
```

##### **2 Tokenisierung/Stemming mit spaCy**

```python
# Tokenisierung (Token = Wort)
import spacy
nlp = spacy.load("de_core_news_sm")   # spaCy-Modell für Deutsch

doc = nlp(text)

tokens = []
# Nur relevante Wortarten, Lemma, keine Stopwörter/Zeichen
for token in doc:
    tokens.append(token)

print(tokens) 
```

```python
# Lemmatisierung
for token in doc:
  print(f'{token}: {token.lemma_}')
```
